# OUTLIER DETECTION

```python
columns_to_drop=['G1','G2','G3','guardian','reason','internet']
data=data.drop(columns_to_drop,axis=1)
```

```python
def detect_outliers(columns):
    outlier_indices = []
    for column in columns:
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1  # IQR is interquartile range.
        mask = (data[column] >= Q1 - 1.5 *
                IQR) & (data[column] <= Q3 + 1.5 * IQR)
        mask = mask.to_numpy()
        false_indices = np.argwhere(~mask)
        outlier_indices.append(false_indices)
    return np.unique(np.concatenate(outlier_indices).ravel())
```

```python
numerical_columns = ['age', 'absences', 'mean_scores']
outlier_indices = detect_outliers(numerical_columns)
print(f'Number of outliers: {len(outlier_indices)}')
```

```python
# Delete outliers
data = data.drop(outlier_indices, axis=0)
data
```

# converting all categorical columns to numerical values

```python
def conv1(r):
    if r == "F":
        return 1
    elif r == "M":
        return 0
data['sex'] = data['sex'] .apply(conv1)
def conv4(r):
    if r == "yes":
        return 1
    elif r == "no":
        return 0
data['romantic'] = data['romantic'] .apply(conv4)
data['schoolsup'] = data['schoolsup'] .apply(conv4)
data['famsup'] = data['famsup'] .apply(conv4)
data['paid'] = data['paid'] .apply(conv4)
data['activities'] = data['activities'] .apply(conv4)
data['nursery'] = data['nursery'] .apply(conv4)
data['higher'] = data['higher'] .apply(conv4)

def conv(r):
    if r == "GP":
        return 1
    elif r == "MS":
        return 2
data['school'] = data['school'] .apply(conv)
def conv2(r):
    if r == "GT3":
        return 1
    elif r == "LE3":
        return 2
data['famsize'] = data['famsize'] .apply(conv2)
def conv3(r):
    if r == "health":
        return 1
    elif r == "teacher":
        return 2
    elif r=='at_home':
        return 3
    elif r=='services':
        return 4
    else:
        return 5
data['Mjob'] = data['Mjob'] .apply(conv3)
data['Fjob'] = data['Fjob'] .apply(conv3)
def conv5(r):
    if r == "U":
        return 1
    elif r == "R":
        return 0
data['address'] = data['address'] .apply(conv5)
def conv6(r):
    if r == "T":
        return 1
    elif r == "A":
        return 2
data['Pstatus'] = data['Pstatus'] .apply(conv6)
```

```python
X, y = data.iloc[:, :-1], data.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.pipeline import Pipeline
```

```python
from sklearn.linear_model import LogisticRegression

#  create model instance 
log= LogisticRegression()

#  Model Fitting
log = log.fit(X_train, y_train)

log_pred = log.predict(X_test)

# Model Accuracy
log_accuracy=accuracy_score(y_test,log_pred)
print("accuracy score:",log_accuracy)

# confusion matrix
print("confusion matrix:\n")
confusion_matrix(y_test, log_pred)
print(confusion_matrix(y_test, log_pred))


#classification report for logistic regression model
print("\nclassification report:\n")
print(classification_report(y_test,log_pred ))
```

# K-NN

```python
# Plotting a accuracy graph
X_axis = list(range(1, 31))
accuracy = pd.Series()

for i in X_axis:
    knn= KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    pred= knn.predict(X_test)
    accuracy= accuracy.append(pd.Series(metrics.accuracy_score(pred,y_test)))
plt.figure(figsize=(9,9))
plt.plot(X_axis, accuracy)  
plt.xticks(range(1,31))
plt.title("Finding best value for n_estimators")
plt.xlabel("n_estimators")
plt.ylabel("Accuracy")
plt.grid()
plt.show()    
print('Highest accuracy: ',accuracy.values.max())
```
